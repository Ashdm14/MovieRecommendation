```python
import ast

def convert(obj):  # this obj is a string
    L = []
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        L.append(i['name'])
    return L 
        
```

```python
def convert3(obj):  # this obj is a string
    L = []
    counter=0
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
        
```

```python
def fetchDirector(obj):  # this obj is a string
    L = []
    for i in ast.literal_eval(obj):  # and in this step it will become a list
        if i['job']=='Director':
            L.append(i['name'])
        break
    return L 
        
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index =new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movie_list =sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    
    return 
```
